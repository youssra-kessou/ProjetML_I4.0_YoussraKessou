**2.1- Gestion des valeurs manquantes**

```python
# Supprimer les valeurs manquantes
data = data.drop(columns=[f'Unnamed: {i}' for i in range(24, 35)])
```

```python
# Imputation des capteurs par la médiane
sensor_cols = ['Current_J0', 'Temperature_T0', 'Current_J1', 'Temperature_J1',
       'Current_J2', 'Temperature_J2', 'Current_J3', 'Temperature_J3',
       'Current_J4', 'Temperature_J4', 'Current_J5', 'Temperature_J5',
       'Speed_J0', 'Speed_J1', 'Speed_J2', 'Speed_J3', 'Speed_J4', 'Speed_J5',
       'Tool_current']
data[sensor_cols] = data[sensor_cols].fillna(data[sensor_cols].median())
```

```python
# Supprimer les lignes où la cible est manquante
data = data.dropna(subset=['Robot_ProtectiveStop'])
```

```python
# Conversion des types
data['Timestamp'] = pd.to_datetime(data['Timestamp'].str.strip('""'), format='%Y-%m-%dT%H:%M:%S.%f%z', errors='coerce')
# Strip extra double quotes and explicitly specify the format
data['grip_lost'] = data['grip_lost'].astype(int)
```

```python
# Sauvegarder le dataset nettoyé
data.to_csv('cobot_data_clean.csv', index=False)
```
